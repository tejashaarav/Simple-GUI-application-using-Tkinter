```python
from tkinter import *
root = Tk()
root.title("hello there!")
root.geometry('350x200')
lbl = Label(root,text = "how are you doing")
lbl.grid()

def clicked():
    lbl.configure(text = "the weather's nice")
btn = Button(root,text = "how's the weather", fg='red',command = clicked)
btn.grid(column=1, row=0)
root.mainloop()
```

```python
from tkinter import *
root = Tk()
root.title("hello there!")
root.geometry('350x200')
lbl = Label(root,text = "how are you doing")
lbl.grid()
txt = Entry(root, width=10)
txt.grid(column =1, row =0)
def clicked():
    res = "You wrote" + " "+ txt.get()
    lbl.configure(text = res)
btn = Button(root,text = "click", fg='red',command = clicked)
btn.grid(column=2, row=0)
root.mainloop()
```

```python
from tkinter import *
root = Tk()
root.title("hello there!")
root.geometry('350x200')

menu = Menu(root)
item = Menu(menu)
item.add_command(label='New')
menu.add_cascade(label='File', menu=item)
root.config(menu=menu)
lbl = Label(root,text = "how are you doing")
lbl.grid()
txt = Entry(root, width=10)
txt.grid(column =1, row =0)
def clicked():
    res = "You wrote" + " " +  txt.get()
    lbl.configure(text = res)
btn = Button(root,text = "click", fg='red',command = clicked)
btn.grid(column=2, row=0)
root.mainloop()
```
